# Keywords Sentiment

```python
keywords_url = 'https://raw.githubusercontent.com/wal-03/Intern---Sentiment-Tracer-PK/refs/heads/main/keywords_verticaly.txt'
```

```python
# Fungsi untuk membaca file keywords dalam format vertikal atau campuran
def fetch_keywords(url):
    response = requests.get(url)
    if response.status_code == 200:
        lines = response.text.strip().split("\n")
        keywords = {}
        current_category = None
        for line in lines:
            line = line.strip()
            if line.startswith("[") and line.endswith("]"):  # Format vertikal
                current_category = line[1:-1]
                keywords[current_category] = []
            elif ":" in line:  # Format horizontal
                current_category, words = line.split(":")
                current_category = current_category.strip()
                keywords[current_category] = [word.strip() for word in words.split(",") if word.strip()]
            elif current_category and line:  # Format vertikal kata kunci
                keywords[current_category].append(line)
        return keywords
    else:
        raise Exception(f"Failed to fetch keywords. Status code: {response.status_code}")


# Ambil dan proses keywords
keywords = fetch_keywords(keywords_url)
keywords
```

# Kategorisasi

```python
# Fungsi untuk mengkategorikan komentar
def categorize_comment(comment, keywords):
    for category, words in keywords.items():
        # Cari apakah ada kata atau frasa yang cocok dalam komentar
        if any(word in comment.lower() for word in words):
            return category
    return 'Uncategorized'

```

```python
# Tambahkan kolom kategori
df['category'] = df['text'].apply(lambda x: categorize_comment(x, keywords))
```

```python
df
```

# Hasil Kategori

```python
# Fungsi untuk menampilkan komentar berdasarkan kategori
def filter_by_category(df, category):
    return df[df['category'] == category]
```

### Positif

```python
# Contoh: Menampilkan komentar dengan kategori "Positif"
positive_comments = filter_by_category(df, "Positif")
print("Komentar dengan kategori Positif:")
positive_comments
```

### Negative

```python
# Contoh: Menampilkan komentar dengan kategori "Negatif"
negative_comments = filter_by_category(df, "Negatif")
print("\nKomentar dengan kategori Negatif:")
negative_comments
```

### Neutral

```python
# Contoh: Menampilkan komentar dengan kategori "neutral"
neutral_comments = filter_by_category(df, "Neutral")
print("\nKomentar Neutral:")
neutral_comments
```

### Uncategorized

```python
# Contoh: Menampilkan komentar dengan kategori "Uncategorized"
uncategorized_comments = filter_by_category(df, "Uncategorized")
print("\nKomentar Uncategorized:")
uncategorized_comments
```
