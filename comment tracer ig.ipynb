# Keywords Sentiment

```python
keywords = {
    "Positif" : ['keren', 'luar biasa', 'mantap', 'mantul', 'thanks', 'suka banget'],
    "Negatif" : ['ilang', 'kepotong', 'diretas']
}
```

# Kategorisasi

```python
# Fungsi untuk mengkategorikan komentar
def categorize_comment(comment, keywords):
    for category, words in keywords.items():
        # Cari apakah ada kata atau frasa yang cocok dalam komentar
        if any(word in comment.lower() for word in words):
            return category
    return 'Uncategorized'

```

```python
# Tambahkan kolom kategori
df['category'] = df['text'].apply(lambda x: categorize_comment(x, keywords))
```
